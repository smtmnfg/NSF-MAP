```python
df=merged6
# Create a new column to store the continuous values
df['Cycle_State_New'] = df['Q_Cell_CycleCount']

# Initialize the offset
offset = 0
count=0
# Loop through the DataFrame and adjust the values
for i in range(1, len(df)):
    if df.loc[i, 'Q_Cell_CycleCount'] == 0:
        offset = df.loc[i - 1, 'Cycle_State_New']
        count+=1
        df.loc[i, 'Description'] = 'E_STOPPED' 
    df.loc[i, 'Cycle_State_New'] += offset  
    
df
```

```python
import pandas as pd

# Initialize the 'actual_state' column with 'normal'
df['actual_state'] = 'normal'

# Define a function to update actual_state based on conditions
def update_actual_state(group):
    anomaly = group['Description'].iloc[0]
    
    if anomaly == 'NoNose':
        group.loc[group['CycleState'].between(8, 21), 'actual_state'] = 'NoNose'
    
    elif anomaly == 'NoNose,NoBody2':
        group.loc[group['CycleState'].between(6, 8), 'actual_state'] = 'NoBody2'
        group.loc[group['CycleState'].between(8, 21), 'actual_state'] = 'NoNose,NoBody2'
        
    elif anomaly == 'NoNose,NoBody2,NoBody1':
        group.loc[group['CycleState'].between(4, 6), 'actual_state'] = 'NoBody1'
        group.loc[group['CycleState'].between(6, 8), 'actual_state'] = 'NoBody2,NoBody1'
        group.loc[group['CycleState'].between(8, 21), 'actual_state'] = 'NoNose,NoBody2,NoBody1'
        
    elif anomaly == 'E-STOPPED':
        group['actual_state'] = 'E-STOPPED'
    
    return group

# Apply the function to each group based on 'cycle_count'
df = df.groupby('Cycle_State_New').apply(update_actual_state)

# Display the updated DataFrame
df
```
